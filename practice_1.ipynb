Stacks: Data is stored in the same order which they arrive, ordered list with insertion and deletion done on the top. LIFO/ FILO. PUSH and POP.
Empty stack POP - Underflow.
Push into Full stack - Overflow.

```python
class Stack:
  def __init__(self, n):
    self.hold = []*n
    self.limit = n

  def push(self, value):
    if len(self.hold) < self.limit:
      self.hold.append(value)
    else:
      raise OverflowError('Stack is full')
      return self.hold

  def pop(self):
    if len(self.hold) == 0:
      raise ValueError('Stack is empty')

    return self.hold.pop()

  def top(self):
    if len(self.hold) == 0:
      raise ValueError('Stack is empty')
    return self.hold[-1]

  def size(self):
    return len(self.hold)

  def isEmpty(self):
    return self.hold == []

  def fullStack(self):
    return len(self.hold) == self.limit

  def __repr__(self):
    return f'The list is {self.hold}'


```

```python
def valid_string(s:str):
  stack = Stack(len(s))
  for i in s:
    if i in '([{':
      stack.push(i)

    elif i in ')]}':
      p = stack.pop()
      if p == '(' and i == ')':
        continue
      elif p == '[' and i == ']':
        continue
      elif p == '{' and i == '}':
        continue
      else:
        return False
    else:
      return 'Not a string'
```

```python
class myQueue:
  def __init__(self, size):
    self.size =  size
    self.queue = [None]*size
    self.head = 0
    self.tail = 0
    self.count = 0

  def enqueue(self, value):
    if self.count < self.size:
      self.queue[self.tail] = value
      self.tail = (self.tail + 1) % self.size
      self.count += 1
    else:
      raise OverflowError('Queue is full')

  def dequeue(self):
    if len(self.queue) == 0:
      raise ValueError('Queue is empty')
    item = self.queue[self.head]
    self.queue[self.head] = None
    self.head = (self.head + 1) % self.size
    self.count -= 1

    return item

  def isEmpty(self):
    return self.queue == []

  def __repr__(self):
    return f'The queue is {self.queue}'
```

```python
def insertion_sort(l):
  '''The input is a unsorted list and the output is a sorted list / array like structure
  1. The first element is sorted
  2. The first two,
  3. The first 3 etc:
  Check the empty list:
  compare the first and last elements else loop through the entire r list
  '''
  if not l:
    return []
  r = []
  r.append(l[0])
  for j in range(1,len(l)):
    key = l[j]
    if key >= r[-1]:
        r.append(key)
        continue
    else:
      for i in range(len(r)):

        if key < r[i]:
          r.insert(i, key)
          break
  return r

```

```python
def merge_sort(l:List):
  '''Input is a list and the output is a sorted version of the list.'''
  if len(l) <= 1:
    return l

  l1 = merge_sort(l[:len(l)//2])
  l2 = merge_sort(l[len(l)//2:])
  return merge(l1,l2)

def merge(l1:List, l2:List):
  '''O(n) operation'''
  result = []
  i = j = 0
  while i < len(l1) and j < len(l2):
    if l1[i] < l2[j]:
      result.append(l1[i])
      i += 1
    else:
      result.append(l2[j])
      j += 1
  result.extend(l1[i:])
  result.extend(l2[j:])
  return result



```

```python
def quick_sort(l, s, e):
  '''This is not usual - it involves swapping and inplace too nlogn'''
  if s >= e:
    return l
  left = s
  for i in range(s,e):
    if l[i] < l[e]:
      l[left],l[i] = l[i], l[left]
      left += 1
  l[left],l[e] = l[e],l[left]
  quick_sort(l,s,left-1)
  quick_sort(l,left+1,e)
  return l
```

```python
def counting_sort(l):
  m = max(l)
  tmp = [0]*(m+1)
  for i in l:
    tmp[i] += 1
  for i in range(1,len(tmp)):
    tmp[i] = tmp[i] + tmp[i-1]
  result = [0]*len(l)
  for i in range(len(l)-1, -1, -1):
    val = l[i]
    position = tmp[val]
    result[position - 1] = val
    tmp[val] -= 1
  return result
```

```python
def binary_search(l, item, left,right):
  '''Assume the list is sorted'''
  if left > right:
    return -1
  guess = (left + right)//2
  center = l[guess]
  if center > item:
    return binary_search(l, item,left,guess-1)
  if center < item:
    return binary_search(l, item, guess+1, right)
  else:
    return guess
```

```python
def binary_search(l,item):

  '''Efficient method of binary search -
  use two pointers to play with the list
  manipulate in place 0log(n)'''
  left = 0
  right = len(l) - 1
  while left <= right:
    mid = (left+right)//2
    if item > l[mid]:
      left = mid + 1
    elif item <l[mid]:
      right = mid - 1
    else:
      return mid
  return -1
```
